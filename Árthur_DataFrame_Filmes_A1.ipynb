# Criando o dicionário e o dataframe

```python
import pandas as pd

# Dicionário com os dados de todos os filmes
dados_filmes = {
    'Título': ['O Senhor dos Anéis', 'Clube da Luta', 'Parasita', 'Interestelar', 'Matrix', 'O Grande Hotel Budapeste'],
    'Diretor': ['Peter Jackson', 'David Fincher', 'Bong Joon-ho', 'Christopher Nolan', 'Lana Wachowski, Lilly Wachowski', 'Wes Anderson'],
    'Ano de Lançamento': [2001, 1999, 2019, 2014, 1999, 2014],
    'Gênero': ['Aventura, Fantasia', 'Drama', 'Comédia, Drama, Suspense', 'Aventura, Drama, Ficção Científica', 'Ação, Ficção Científica', 'Comédia, Crime'],
    'Duração (min)': [178, 139, 132, 169, 136, 200],
    'Avaliação IMDB': [8.9, 8.8, 8.6, 8.6, 8.7, 8.1],
    'Receita (em milhões)': [134.97, 101.2, 258.8, 677.5, 465.3, 105.4]
}

# Criando o DataFrame
df_filmes = pd.DataFrame(dados_filmes)
```

```python
display(df_filmes)
```

# 7. Adicione um item (linha) na sua base de dados.

```python
print("Adicionando um novo registro:")
novo_filme = {
    'Título': 'Inception',
    'Diretor': 'Christopher Nolan',
    'Ano de Lançamento': 2010,
    'Gênero': 'Action, Adventure, Sci-Fi',
    'Duração (min)': 148,
    'Avaliação IMDB': 8.8,
    'Receita Bruta (milhões)': 829.9
}

df_novo_filme = pd.DataFrame([novo_filme]) # Convertendo o dicionário em dataframe
display(pd.concat([df_filmes, df_novo_filme], ignore_index=True)) # Concatenando ambos dataframes
```
